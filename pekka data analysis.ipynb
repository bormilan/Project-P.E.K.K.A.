```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Video game chat questions(1-52).xlsx')
```

```python
data_useful = data[[
    'ID','Age','Gender',','Hours spent playing video games on a weekly basis:',
    'What type of video games do you play ?','How often do you encounter toxicity on a scale of 0 to 10?','How much does toxic behaviour bother you on a scale of 0 to 10?',
    'Have you ever stopped playing because of overly toxic playmates?','Place the categories in order of how often you experience them in-game.','Have you ever gotten mad while playing a game ?',
    'Game mechanics','Outcome of the game','Your bad performance','Teammates\' bad performance','Opponnents\' good performance','Experience gap between your teammates',
    'Experience gap between your team and opponnent team','Teammates\' behaviour','Opponents\' behaviour','Is there any other reason you have gotten mad while playing an online game ?',
    'Do you think chat use should be restricted to well-behaving players only?','Do you think players would be less toxic if the punishments were harsher?','Can a toxic player reform and become a constructing part of the community?',
    'Is a mute function enough for you to completely close out toxicity?','Please put the following situations in order of how annoying they are for you!','Do you prefer video games where communication is restricted/turned off entirely?\nIs mic-talk a bad addition to the game?',
    'If you answered one of the first two answers in the above question (you don\'t like the idea of a chatbox), what was your reasoning for it?','Have you ever made friends or enemies in a video game?','Do you prefer voice chat or text chat?'
]]
```

```python
data_useful.shape
```

```python
yes_stopped = data_useful.loc[(data_useful['Have you ever stopped playing because of overly toxic playmates?'] == 'Yes'),['Have you ever stopped playing because of overly toxic playmates?']].count()
no_stopped = data_useful.loc[(data_useful['Have you ever stopped playing because of overly toxic playmates?'] == 'No'),['Have you ever stopped playing because of overly toxic playmates?']].count()
```

```python
labels = 'Yes','No'
sizes = [int(yes_stopped),int(no_stopped)]
explode = ( 0.1, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=['#20B2AA','#CF4449'])
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.savefig('stopped_playing.png')
```
